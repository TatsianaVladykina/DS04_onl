```python
#8)  Задание: Посчитать четные и нечетные цифры числа
#Определить, сколько в числе четных цифр, а сколько нечетных. Число вводится с клавиатуры.

n = int(input())
count_ch = 0
count_nech = 0
while n > 0:
  if n % 2 == 0:
    count_ch += 1
  else:
    count_nech += 1
  n //= 10    
print("Четных числел: ", count_ch)  
print("Нечетных числел: ", count_nech)
```

```python
#9) Циклический сдвиг

a = [3,4,5,6,7,8,9]
def shift(a,step):
    if step < 0:
        for i in range(abs(step)):
            a.append(a.pop(0))
    else:
        for i in range(step):
            a.insert(0,a.pop())

shift(a, 3)
print(a)
```
