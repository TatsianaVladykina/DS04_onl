1.	Как бы вы определили машинное обучение?

Машинное обучение - это область искусственного интеллекта, которая использует алгоритмы и статистические методы для анализа данных, определения закономерностей и построения моделей, которые могут выполнять предсказания или действовать в некоторых других определенных контекстах.

2.	Можете ли вы назвать три типа задач в которых машинное обучение показывает хорошие результаты.
Машинное обучение часто применяется в следующих типах задач:

Классификация: определение категории объектов, например, определение эмоциональной окраски текста.
Регрессия: предсказание численных значений, например, предсказание цены акции.
Кластеризация: разделение объектов на группы на основе их похожести.




3.	Что такое помеченный обучающий набор? В каких моделях используются такие наборы?
Помеченный обучающий набор - это набор данных, где каждый элемент данных имеет связанный с ним метку, которая является искомым значением. Эти метки используются для того, чтобы модель машинного обучения могла изучить закономерности в данных и применить эти закономерности для предсказания значений для неизвестных данных.

Помеченные обучающие наборы используются в задачах классификации и регрессии. Например, в задаче классификации может использоваться помеченный обучающий набор, где каждый элемент данных содержит изображение объекта, а метка определяет класс, к которому он относится. В задаче регрессии помеченный обучающий набор может содержать информацию о времени и тем

4.	Какой тип алгоритма машинного обучения вы бы использовали для обучения робота прохождению неизведанной территории?
Для обучения робота прохождению неизведанной территории, возможно, был бы подходящим использовать алгоритм машинного обучения основанный на методе обучения с подкреплением (Reinforcement Learning). Этот алгоритм позволяет роботу изучать окружающую среду, получать определенные награды или штрафы в зависимости от его действий и исправлять свои действия в соответствии с полученными опытом и оценками.

5.	Какой тип алгоритма вы бы применили для разделения своих заказчиков на несколько групп.
Для разделения заказчиков на несколько групп можно использовать алгоритм кластеризации. Он основан на поиске закономерностей в данных, и позволяет группировать объекты, которые похожи друг на друга, в одну категорию. 

6.	Как вы думаете - задача фильтрации спама, это задача обучения с учителем или без учителя?
Задача фильтрации спама обычно является задачей обучения с учителем. Это означает, что у нас есть помеченные данные - сообщения, которые мы знаем, являются спамом или нет, и мы используем эту информацию для обучения модели классификации. Модель затем может использоваться для фильтрации новых, непомеченных сообщений.

7.	Назовите основные проблемы машинного обучения?
Недостаточный набор данных: недостаточный набор данных может привести к недообучению модели, что в свою очередь приведет к низкой точности предсказаний.

Переобучение: переобучение происходит, когда модель слишком подстраивается под обучающие данные, поэтому она не способна эффективно работать с новыми данными.

Неравномерное распределение классов: неравномерное распределение классов может привести к проблемам с предсказанием, поскольку модель может недооценить некоторые классы.

Несоответствие модели и задачи: использование неподходящей модели для решения определенной задачи может привести к низкой точности предсказаний.

Сложность выбора метрики: выбор неподходящей метрики для оценки

8.	Что происходит, если ваша модель хорошо работает на обучающем наборе данных и плохо обобщается на новые образцы? Предложите варианты решения проблемы.

Это называется переобучением. Когда модель хорошо работает на обучающем наборе данных, но плохо обобщается на новые данные, это говорит о том, что она запомнила и изображает слишком много характеристик обучающего набора данных, включая шум или выбросы, вместо того, чтобы выявлять основные закономерности.

Следующие варианты могут быть использованы для решения проблемы переобучения:

1/Увеличение количества данных в обучающем наборе данных.

2/Использование регуляризации для ограничения сложности модели.

3/Отбор признаков для улучшения качества модели.

4/Использование перекрестной проверки или использование валидационного набора данных для оценки качества модели.
