```python
import numpy as np
```

```python
data = np.loadtxt('linreg_4.txt')
```

```python
data.shape
```

```python
X_ = data[:, :-1]  # Все колонки, кроме последней
# Добавление столбца единиц (для удобства вычислений)
X = np.c_[np.ones((100,1)), X_]
X
```

```python
y_ = data[:, -1]   # Последняя колонка
y = y_.reshape(100,1)
y
```

```python
# устанавливаем значения для lr и количества итераций
lr = 0.1
iters = 100
m = 100

# инициализируем начальные значения параметров
h = np.random.randn(4,1)
# применяем градиентный спуск
for i in range(iters):
    gradient = (2/m)*X.T @ (X @ h - y) 
    h = h - lr * gradient
```
