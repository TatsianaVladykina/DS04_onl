Из каждой тренировочной папки выберем случайным образом 10% файлов, скопируем их в validation_data и удалим их из исходной папки.

```python
for i in range(len(train_dirs)):
    fnames = os.listdir(train_dirs[i])
    toCopy = len(fnames)//10

    filesToCopy = []
    for j in range(toCopy):
        filesToCopy.append(rn.choice(fnames))
    
    # cOPY
         
    for fi in filesToCopy:
        src = os.path.join(train_dirs[i],fi)
        dst = os.path.join(test_dirs[i], fi)
        shutil.copyfile(src,dst)

    
    for fi in filesToCopy:
        try:
            src = os.path.join(train_dirs[i],fi)
            os.remove(src)
            print(src+' deleted')
            #time.sleep(1)
        except:
            pass
```

## подготовка генераторов

```python
#Подготовим генератор изображений, который будет извлекать изображения из размеченных папок,
#приводить их к нужному размеру и готовить из них батчи.
Pic_Y, Pic_X = 126, 224
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255)
                            
train_generator = train_datagen.flow_from_directory('/content/FORLEARNFULL/to_train',
                                                   target_size = (Pic_Y, Pic_X),
                                                   batch_size = 64,
                                                   class_mode = 'categorical')
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
                            
test_generator = test_datagen.flow_from_directory('/content/FORLEARNFULL/to_test',
                                                   target_size = (Pic_Y, Pic_X),
                                                   batch_size = 64,
                                                   class_mode = 'categorical')
```

```python
# Обучение модели с 0

model = models.Sequential()

model.add(layers.Conv2D(filters=32, kernel_size=3, activation="relu", input_shape = (Pic_Y, Pic_X, 3)))
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Conv2D(filters=64, kernel_size=3, activation="relu"))
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Conv2D(filters=128, kernel_size=3, activation="relu"))
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Conv2D(filters=256, kernel_size=3, activation="relu"))
model.add(layers.MaxPooling2D(pool_size=2))
model.add(layers.Conv2D(filters=256, kernel_size=3, activation="relu"))

model.add(layers.Flatten())
model.add(layers.Dense(9, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",
              optimizer="rmsprop",
              metrics=["accuracy"])
```

```python
callbacks = [tf.keras.callbacks.ModelCheckpoint(
        filepath="convnet_from_scratch.h5",
        save_best_only=True,
        monitor="val_loss")]
```

```python
history = model.fit(train_generator, epochs=10, validation_data=test_generator, callbacks=callbacks)
```

```python
import numpy as np
from tensorflow import keras

def predict_image(image_path):
    # Загрузка изображения и подготовка его для подачи в модель
    img = keras.preprocessing.image.load_img(image_path, target_size=(Pic_Y, Pic_X))
    img = keras.preprocessing.image.img_to_array(img)
    img = img / 255.0  # Нормализация значений пикселей
    img = np.expand_dims(img, axis=0)

    # Предсказание класса
    predicted_probs = model.predict(img)
    predicted_class = np.argmax(predicted_probs)
    
    return predicted_class
```

**Перенос знаний!**
Предобученная модель VGG_FIX_model.h5

```python
# загрузка сверточных слоев 

conv_base = tf.keras.applications.vgg16.VGG16(
    weights="imagenet",
    include_top=False,
    input_shape=(126, 224, 3))
```

```python
# Заморозка сверточных слоев
conv_base.trainable = False #изменять веса не надо
```

```python
# Конструируем модель
model_3 = models.Sequential()
model_3.add(conv_base)
model_3.add(layers.Flatten())
model_3.add(layers.Dense(9, activation = 'softmax'))
```

```python
model_3.compile(loss="categorical_crossentropy",
              optimizer="rmsprop",
              metrics=["accuracy"])
```

```python
callbacks = [tf.keras.callbacks.ModelCheckpoint(
        filepath="VGG_FIX_model.h5",
        save_best_only=True,
        monitor="val_loss")]
```

```python
model_3.summary()
```

```python
# фитим на тех же генераторах то есть с расширением данных
history = model_3.fit(train_generator, epochs=5, validation_data=test_generator, 
    callbacks=callbacks)
```
