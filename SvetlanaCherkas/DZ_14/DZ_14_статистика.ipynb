```python
# mean - среднее арифметическое всех значений в выборке.  экстремально большие или маленькие значения оказывают сильное влияние на арифметическое среднее
# у клиентов из Германии средний возраст почти совпадает с возрастом клиентов по всей выборке (39.8 и 38,9)
# а вот баланс средств на счету (119730.1) у клиентов из Германии гараздо выше, чем в целом по выборке (76485.9)
```

```python
# median - это значение, которое делит упорядоченный набор данных на две равные половины.
# медиана считается более “робастной” оценкой, т.е. более устойчивой к выбросам и крайним значениям.
# у клиентов из Германии среднее арифметическое значений  и медиана баланса средств на счету почти равна, это
# нам говорит о том, что данные не имеют больших выбросов и крайних значений.
```

```python
# Стандартное отклонение - это мера разброса значений в наборе данных относительно их среднего значения.
# у клиентов из Германии стандартное отклонение возраста совпадает со стандартным отклонением возраста клиентов по всей выборке (10.5)
# а вот стандартное отклонение баланса на счету клиентов из Германии намного меньше, чем по всей выборке,
# что говорит о маленьком разбросе значений по клинтам банка из Германии 
```
