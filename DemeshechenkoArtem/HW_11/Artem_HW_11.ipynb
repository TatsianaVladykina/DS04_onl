```python
#2 отделяем целевую переменную (последнее значение в каждом векторе)
Y = data[:, 3:4] # срез всех строк и последнего столбца
Y # целевая переменная
```

```python
# отделяем независимые переменные и добавляем смещение (np.ones)
X = np.c_[np.ones((data.shape[0], 1)), data[:, 0:3]] # независимые переменные
```

```python
# 3 и 4
lr = 0.01 # скорость обучения
iters = 1000 # кол-во шагов 
m = data.shape[0] # кол-во образцов (получается кол-во наших векторов (строк))

h = np.random.randn(4, 1) # инициализация вектора h, 4 значения h0, h1, h2, h3

# поиск
for i in range(iters):
    gradient = (2/m)*X.T @ (X @ h - Y)
    h = h - lr * gradient

print(h) # y = h0 + h1x1 + h2x2 + h3*x3 
```
