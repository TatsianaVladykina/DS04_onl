easy lvl: декоратор разворачивающий аргументы
Пример:a(1,2,3) # > a(3,2,1)
```
def rev(func):

  def wrapper(*args, **kwargs):
    sp = func(*args, **kwargs)
    print(sp[::-1])
  return wrapper

@rev
def function(a,b,c):
  return a, b, c

function(1,2,3)
```
medium lvl: декоратор фильтрующий указаный элемент

Пример: 
```
@my_filter(filt="Oleg")
def a(g,h,j):
    pass

a(1,2,"Oleg") # > Exception
```
Решение:
```
def filt(filter):

  def rev(func):

    def wrapper(*args, **kwargs):
      sp = func(*args, **kwargs)
      result = []
      for i in sp:
        if i != filter:
          result.append(i)
      return result
    
    return wrapper
  
  return rev

@filt("Luk")
def function(a, b, c):
  return a, b, c

print(function(1,2,"Luk"))
```

hard: кеширующий декоратор
пример:
```
cashe = {}

@cashe_func
def f(a,b,c):
    return a * b * c

f(1,2,3) # > calculte
f(1,2,3) # > result from cashe
```
Решение:
```
def hasher(func):

  def wrapper(*args, **kwargs):
    sp = func(*args, **kwargs)
    
    return hash(sp)
    
  return wrapper

@hasher
def function(a, b, c):
  return a * b * c

print(function(1, 2, 3))
```
