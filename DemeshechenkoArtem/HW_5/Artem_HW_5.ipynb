Задание 1: Числа Фибоначчи
Числа Фибоначчи – это ряд чисел, в котором каждое следующее число равно сумме двух предыдущих: 1, 1, 2, 3, 5, 8, 13, ... . Иногда ряд начинают с нуля: 0, 1, 1, 2, 3, 5, ... . В данном случае мы будем придерживаться первого варианта.

Формула:
F1 = 1
F2 = 1
Fn = Fn-1 + Fn-2
Пример вычисления:
F3 = F2 + F1 = 1 + 1 = 2
F4 = F3 + F2 = 2 + 1 = 3
F5 = F4 + F3 = 3 + 2 = 5
F6 = F5 + F4 = 5 + 3 = 8
```
def fib(a):
  if a in (1, 2):
    return 1
  return fib(a-1) + fib(a-2)

print(fib(9))
```

def bank(a, years):
  if years >= 1:
    a += a * 0.1

def col(n: int):
  sp: int = 1

    if n % 2 == 0:

````python
from math import sqrt
  
def is_prime(n):
  i = 2
  while i <= sqrt(n):
    if n % i == 0:
      return False
    i += 1
  if n > 1:
    return True

def dell(a):
  dell_list = [0]
  for i in range(a):
    try:
      if a % i == 0:
        dell_list.append(i)
    except ZeroDivisionError:
      continue
  return dell_list

max = 0
a = int(input())#600851475143

mass = dell(a)
for i in mass:
  if is_prime(i) and max < i:
    max = i
print("max:", max)
````
