Написать функцию расёта обратной матрицы.

````python
import numpy as np
````

````python
def inv1(matrix):
  if matrix.shape[0] == matrix.shape[1]:
    det_matrix = np.linalg.det(matrix)
    if det_matrix:
      print(matrix, "\n")
      matrix_M = matrix.copy()
      for i in range(0, len(matrix)):
        for j in range(0, len(matrix)):
          matrix_copy = np.delete(matrix, i, axis = 0)
          matrix_copy = np.delete(matrix_copy, j, axis = 1)
          matrix_M[i][j] = round(np.linalg.det(matrix_copy) * pow(-1, i+j))
      matrix_inv = matrix_M.T * 1/det_matrix
      return matrix_inv
    else:
      print("Вырожденная матрица")
      return matrix
  else:
      print("Не квадратная матрица")
      return matrix

A = np.random.randint(-5, 6,(3, 3))
print(inv1(A))
print()
print(np.linalg.inv(A))
````
