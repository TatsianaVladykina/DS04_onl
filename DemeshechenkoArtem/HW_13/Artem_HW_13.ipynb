```python
import math
```

```python
# размещение
def C(n, m):
  return int(math.factorial(n) / (math.factorial(n-m) * math.factorial(m)))
```

```python
# сочетание
def A(n, m):
  return int(math.factorial(n) / (math.factorial(n-m)))
```

```python
# перестановка
def P(n):
  return math.factorial(n)
```

```python
# 1
print(A(23, 2))
```
