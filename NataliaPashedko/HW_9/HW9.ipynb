```python
a = np.linspace(1, 9, 9)
print(a)
b = a[::-1]
print(b)
```

```python
print(a < b)
```

```python
print(a == b)
```

```python
print(a > b)
```

```python
c = a * b
print(c)
```

```python
d = c.mean()
print(d)
```
