```python
import os

input_word = input("Введите слово для поиска в книгах библиотеки: ").split()

result_set = []  # список названий книг, в тексте которых есть введенное пользователем слово
errors = []      # список поврежденных книг

for bookname in os.listdir("/content/drive/MyDrive/Colab Notebooks/books"):
   try:
     with open(os.path.join("/content/drive/MyDrive/Colab Notebooks/books", bookname), mode = 'r', encoding = "Windows-1251") as file:
       for word in input_word:
         if word.lower() in file.read().lower():  # приведение слов к одному регистру
          result_set.append(bookname)
   except:
     errors.append(bookname)

if len(result_set) != 0:
  print("\n" "Результаты поиска:")
  for i in result_set:
    print(i)
else:
  print("\n" "По Вашему запросу ничего не найдено.")

if len(errors) != 0:
  print("\n" "Поврежденные файлы:")
  for i in errors:
    print(i)
else:
  print("\n" "Поврежденных файлов не найдено.")
```
