Стратегии решения проблем пустых строк:

````python
hdf.dropna(subset = ['total_bedrooms']) # метод снесет строки с пустыми значениями по заданному столбцу, но в обученную модель могут прийти данные, у которых total_bedrooms тоже None, что тогда? Поможет 3 способ
````

````python
hdf.drop(['total_bedrooms'], axis = 1) # метод снесет столбец целиком => нам не очень подходит этот способ, так как значений не так много (всего лишь 207)
````

````python
mediantb = hdf['total_bedrooms'].median() # ищем медиану и заполняем пустые значения ею (все остальные значения останутся на месте, в том числе и нулевые) 
hdf['total_bedrooms'].fillna(mediantb, inplace = True) # inplace = True означает, что мы меняем сами значения в самом дата фрейме

#medianpbr = hdf['bebrooms_per_room'].median()
#hdf['bebrooms_per_room'].fillna(medianpbr, inplace = True)
````

Новые атрибуты (конструирование новых признаков), так как не все вышеперечисленные признаки информативны


````python
hdf['rooms_per_house'] = hdf['total_rooms'] / hdf['households'] # формируем новые столбцы, но сперва нужно избавиться от пустых значений в столбце total_bedrooms, иначе будут ошибки, мы уже это сделали
hdf['bebrooms_per_room'] = hdf['total_bedrooms'] / hdf['total_rooms']
hdf['population_per_house_hold'] = hdf['population'] / hdf['households']
````

Разделим наши данные на тестовый и проверочный наборы

Создание проверочного набора

````python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train,  Y_test =  train_test_split(X, Y, test_size = 0.2, random_state = 42)
# в функцию train_test_split передаем X (все данные без слов), Y (множество меток), test_size (откладываем 20% в тестовое множество), random_state (в генератор случайных чисел устанавливается значение 42)
# поставив 42 (ничего не значит, но так принято "ха-ха, универсальный ответ на все"), будем получать всегда одинаковое разбиение, при следующем вызове функции будут те же индексы
#то есть random_state - это фиксация внутренних параметров модели (точка старта)
````

````python
from sklearn.linear_model import LinearRegression # импортируем линейную регрессию
linreg = LinearRegression() # создаем объект
````

````python
linreg.fit(X_train, Y_train) # передаем в fit множество признаков и множество меток и линейная регрессия натренируется (сама и за секунду)
````

GRID SEARCH
(здесь подбираются гиперпараметры - числа, которые модель характеризуют)
Если говорим про нейросеть, то говорим о кол-ве нейронов, слоев
Если говорим про случайный лес, то говорим о кол-ве деревьев и т.д.)

````python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_estimators': [3,10,30], 'max_features':[2,4,6,8]} # n_estimators задаем кол-во деревьев (3 вар-та: 3 дерева, 10 и 30), max_features - сколько фич нужно отбираться из сета для оптимального обучения
# сначала 2 признака, потом 4, 6 и 8. Причем какие именно признаки выбирает сама модель

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv = 5, scoring = 'neg_mean_squared_error')
````

````python
grid_search.fit(X_train, Y_train) # модель обучилась
````

````python
grid_search.best_params_ # смотрим лучшие параметры: достаточно использовать 6 фичей (а не все), но деревьев нужно побольше: 30 (в идеале в 'n_estimators': [3,10,30] нужно добавить больше вариантов и запустить еще раз)
````

````python
grid_forest = grid_search.best_estimator_
````

````python
preds = grid_forest.predict(someData)
````

````python
mse = mean_squared_error(someLabels, preds) # снова оцениваем прогноз
rmse = math.sqrt(mse)
rmse # лучший результат
````

На тестовом наборе

````python
preds = grid_forest.predict(X_test)
mse = mean_squared_error(Y_test, preds) # снова оцениваем прогноз
rmse = math.sqrt(mse)
rmse  # плохо, возможно плохой тестовый набор
````
