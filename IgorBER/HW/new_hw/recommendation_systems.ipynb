По логике в этот датасет входили транзакции с минимум 5 продуктами, также есть пустые (не nan значения)-стоят пробелы 

```python
df = df.replace(' ', pd.np.nan)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = df.stack().groupby(level=0).apply(list).apply(lambda x: [i for i in x if str(i) != 'nan'])
```

```python
df
```

```python
encoder = TransactionEncoder()
transactions = pd.DataFrame(encoder.fit(df).transform(df), columns=encoder.columns_)
display(transactions.head())
```

```python
frequent_itemsets = apriori(transactions, min_support= 6/len(df), use_colnames=True, max_len = 3)
rules = association_rules(frequent_itemsets, metric="lift",  min_threshold = 1.4)
display(rules.head(10))
print("Rules identified: ", len(rules))
```

```python
shampoo_rules = rules[rules['consequents'].astype(str).str.contains('cheeses')]
shampoo_rules = shampoo_rules.sort_values(by=['lift'],ascending = [False]).reset_index(drop = True)
display(shampoo_rules.head(20))
```
