```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
X_train = pd.read_csv('GiveMeSomeCredit-training.csv', sep=',')
X_test = pd.read_csv('GiveMeSomeCredit-testing.csv', sep=',')
```

```python
X_train.head()
```

```python
X_train.isna().sum()
```

```python
X_train['MonthlyIncome'] = X_train['MonthlyIncome'].apply(lambda x: X_train['MonthlyIncome'].mean() if pd.isna(x) else x)
```

```python
X_train['NumberOfDependents'] = X_train['NumberOfDependents'].apply(lambda x: X_train['NumberOfDependents'].mean() if pd.isna(x) else x)
```

```python
X_train.isna().sum()
```

```python
X_test.head()
```

```python
X_test.isna().sum()
```

```python
X_test['MonthlyIncome'] = X_test['MonthlyIncome'].apply(lambda x: X_test['MonthlyIncome'].mean() if pd.isna(x) else x)
```

```python
X_test['NumberOfDependents'] = X_test['NumberOfDependents'].apply(lambda x: X_test['NumberOfDependents'].mean() if pd.isna(x) else x)
```

```python
X_test.isna().sum()
```

```python
X_train = X_train.drop(['Unnamed: 0'], axis = 1)
X_test = X_test.drop(['Unnamed: 0'], axis = 1)

X_test = X_test.drop(['SeriousDlqin2yrs'], axis = 1)
```

```python
Y_train = X_train['SeriousDlqin2yrs']
```

```python
X_train = X_train.drop(['SeriousDlqin2yrs'], axis = 1)
```

```python
X_train.head(5)
```

```python
Y_train.head(5)
```

```python
ss = StandardScaler()
```

```python
ss.fit(X_train)
```

```python
X_train = ss.transform(X_train)
```

```python
X_test = ss.transform(X_test)
```

```python
X_train.shape
```

```python
model = models.Sequential()

model.add(layers.Dense(16, activation="softmax", input_shape = (10,)))
model.add(layers.Dense(8, activation="softmax"))
model.add(layers.Dense(1, activation="sigmoid"))
```

```python
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(X_train, Y_train, epochs=10, batch_size=100)
```

```python
predictions = model.predict(X_test)
```
