# **Task 1** (Secure Aunthentication)

**[loops, hash, dict] Integrate the algorithm for hiding login and password into existing code.**


**Important note:** *There is a ready-made code below. Your task is to modify it so that it meets the following requirements.*

**Acceptance criteria:**

1.   Must be implemented: *singin, login*
2.   When using *login* - the session is occupied by the user whose credentials were used to log in.
3.   The user can not log in under another data if he has an active old session.
4.   User data is stored in the database.
5.   Only hashed information should be stored in the database.



```python
# frontend --------------------------
info = """
USER:\t\t{} 
LAST CMD:\t{}
"""

# database --------------------------
database = {}

# session ---------------------------
current_user = 'anonimus'

# backend ---------------------------
working = True
cmd = "null"

while working:
    print(info.format(current_user, cmd))
    cmd = input(">> ")
    # регистрация пользователя
    if cmd == "singin":
        while True:
            login = input("login: ")
            if hash(login) in database:
                print("Такой логин уже существует!")
                continue
            else:
                while True:
                    # проверка правильности повторного ввода пароля
                    password_1 = input("password: ")
                    password_2 = input("repeat password: ")
                    if password_1 != password_2:
                        continue
                    else:
                        break
                break
        # добавляем нового пользователя в базу данных
        database[hash(login)] = hash(password_1)
        # устанавливаем текущую сессию
        current_user = login
        
    elif cmd == "login":
        # занята ли текущая сессия другим пользователем 
        if hash(current_user) in database:
            print('Вы не можете войти в систему, т.к. текущая сессия занята другим пользователем!')
        else:
            # есть ли такой пользователь в системе
            while True:
                login = input("login: ")
                if hash(login) not in database:
                    print('Такого пользователя нет в базе данных!')
                    continue
                else:
                    while True:
                        password = input("password: ")
                        if hash(password) == database.get(hash(login)):
                            current_user = login
                            break
                        else:
                            print('Пароль неверный. Попробуйте еще раз!')
                            continue
                    break
                
    elif cmd == "log out":
        current_user = 'anonimus'

    else:
        working = False
else:
    print("WORK COMPLETED")
```
