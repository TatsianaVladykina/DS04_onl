**ЗАДАНИЕ 6**

Написать функцию программу, принимающую 1 аргумент — число от 0 до 1000, и возвращающую True, если оно простое, и False - иначе

```python
number = int(input('введите ваше число: '))
if number == 0 or number == 1:
    print('ваше число не может быть простым или составным, т.к. это 0 или 1')
else:
    for i in range (2, number): # не делим наше число на 1 и на само число, т.к. 100% делятся
        if number % i == 0: # если делится на какое-то другое число в промежутке - не является простым. выходим из цикла
            print('не является простым числом')
            break
        elif number % i != 0 and i != (number - 1): # число не делится на другое, но цикл еще не закончился. идем на новый круг цикла
            continue
        elif number % i != 0 and i == (number - 1): # число не делится на другое и это был последний элемент проверки
            print('простое число')
```

Посчитать четные и нечетные цифры числа
Определить, сколько в числе четных цифр, а сколько нечетных. Число вводится с клавиатуры.

```python
number = input('введите число: ')
num_list = list(number)
chet = 0
nechet = 0
for i in num_list:
    if int(i) % 2 == 0:
        chet += 1
    else:
        nechet += 1
print('в числе', number, 'четных чисел -', chet, ', а нечетных -', nechet)
```

**ЗАДАНИЕ 9**

Циклический сдвиг
Выполнить циклический сдвиг в списке целых чисел на указанное число шагов.
Сдвиг также должен быть кольцевым, то есть элемент, вышедший за пределы списка, должен появляться с другого его конца.
Для решения этой задачи можно воспользоваться следующими методами списка:
append() - добавляет элемент в конец списка,
insert() - вставляет элемент по указанному индексу,
pop() - извлекает элемент с конца списка или, если был передан индекс, по индексу.
Пусть функция shift() в качестве аргументов принимает список и количество шагов сдвига.
Если шаги представлены положительным числом, то сдвиг выполняется вправо, то есть надо извлечь последний элемент и добавить его в начало.
Если шаги - отрицательное число, то будем выполнять сдвиг справа налево, то есть надо извлечь первый элемент и добавить его в конец.

```python
shag = int(input('введите шаг циклического сдвига: '))
spis = [1, 2, 3, 4]
print('первоначальный список:' , spis)
if shag < 0:
    for i in range(abs(shag)):
        spis.append(spis.pop(0))
else:
    for i in range(shag):
        spis.insert(0, spis.pop(-1))
print ('финальный список:' ,spis)
```

**Банкомат**

```python
bankomat = {
    200 : 10,
    100 : 34,
    50 : 32,
    20 : 29,
    10 : 26,
    5 : 15
}
summa = int(input('введите нужную сумму денег: '))
n_200 = summa // 200
summa -= n_200 * 200
n_100 = summa // 100
summa -= n_100 * 100
n_50 = summa // 50
summa -= n_50 * 50
n_20 = summa // 20
summa -= n_20 * 20
n_10 = summa // 10
summa -= n_10 * 10
n_5 = summa // 5
summa -= n_5 * 5
# сколько выдаем
vidacha ={}
vidacha[200] = n_200
vidacha[100] = n_100
vidacha[50] = n_50
vidacha[20] = n_20
vidacha[10] = n_10
vidacha[5] = n_5
print('выдача денег:', vidacha)
# изменение остатка в банкомате
bankomat[200] = bankomat.get(200) - n_200
bankomat[100] = bankomat.get(100) - n_100
bankomat[50] = bankomat.get(50) - n_50
bankomat[20] = bankomat.get(20) - n_20
bankomat[10] = bankomat.get(10) - n_10
bankomat[5] = bankomat.get(5) - n_5
print ('остаток в банкомате:',bankomat)
```
