Напишите программу, которая запрашивала бы у пользователя два числа и выводила бы True или False в зависимости от того, больше первое число второго или нет.

```python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print(a < b)
```

Пользователь вводит трехзначное число. Программа должна сложить цифры, из которых состоит это число. Например, если было введено 349, программа должна вывести на экран число 16, так как 3 + 4 + 9 = 16.

```python
number = int(input("Enter the number: "))
a = number % 10 #получаем третью цифру
number2 = number // 10 
b = number2 % 10 #получаем вторую цифру
number3 = number2 // 10 
c = number3 % 10 #получаем третью цифру
sum = a + b + c
print(sum)
```
