Пример:a(1,2,3) # > a(3,2,1)

```python
def rezvorot(func):
  def qwe():
    a = func()
    return a[::-1]
  return qwe

@rezvorot
def f():
  s = []
  for i in range(10):
    s.append(i)
  return s

f()
```

```python
import random
class Answers:

  def __init__(self, question_num):
    if question_num == 1:
      self.question = 'Откуда на Беларусь готовилось нападение?'
      self.answers_list = ['С планеты Нибиру.', 'Откуда мне знать?', 'там 4 позиции, я карту принёс, сейчас покажу', 'Со склада грязи.', 'С загнивающег запада.']
      self.correct = 'там 4 позиции, я карту принёс, сейчас покажу'
    
    elif question_num == 2:
      self.question = 'Что такое 134340?'
      self.answers_list = ['Числовое обозначение Плутона', 'Количество видов насекомых ', 'Население Мальты', 'Протяженность самой большой дороги в мире', 'Средняя зарплата в РБ (в долларах)']
      self.correct = 'Числовое обозначение Плутона'
    
    elif question_num == 3:
      self.question = 'Какую еду Мстители идут есть после битвы за Нью-Йорк в первом фильме о Мстителях по предложению Тони Старка?'
      self.answers_list = ['Шаурма', 'Бургеры', 'Вкусно и точка', 'Мороженое', 'Пицца']
      self.correct = 'Шаурма'
    print('Вопрос:', self.question)

  # выбираем вопрос
  def dice_roll(self):
    num = random.randint(1, self.sides)
    print('Выпало значение: ', num)
    return num

  # перемешивание списка ответов
  def shuffle_list(self):
    random.shuffle(self.answers_list)
    for i in range(len(self.answers_list)):
      print(i, ' - ', self.answers_list[i])
    return self.answers_list
  
  # проверка ответов
  def check_ans(self, answ):
    if self.answers_list[answ] == self.correct:
      print('Ответ верный')
    else:
      print('Ответ неверный!')

# ---------
class Dice:

  def __init__(self):
    self.sides = 3
  
  def dice_roll(self):
    num = random.randint(1, self.sides)
    print('Выпал вопрос номер: ', num)
    return num

cube = Dice()
question = cube.dice_roll() # рандом вопроса

q1 = Answers(question)
b = q1.shuffle_list() # перемешиваем ответы
ans = int(input('Введите ваш ответ: '))
q1.check_ans(ans)
```
