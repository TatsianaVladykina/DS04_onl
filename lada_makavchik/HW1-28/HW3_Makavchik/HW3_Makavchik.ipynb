```python
import os

word = input('введите слово, которое хотите найти в книгах: ')

for (root,dirs,files) in os.walk('/content/drive/MyDrive/Colab Notebooks/books'):
  print (files)
  for i in files:
    try:
      with open('/content/drive/MyDrive/Colab Notebooks/books/' + i,'r', encoding='windows-1251') as my_file:
        for line in my_file.readlines():
          if word in line:
            print(i)
            break
    except:
      continue
```
