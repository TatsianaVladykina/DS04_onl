1. (СДЕЛАЛА) В списке чисел проверить, все ли элементы являются уникальными, т. е. каждое число встречается только один раз.

[9, 47, 78, 54, 92, 63, 59, 89, 96, 47, 95, 29, 99, 20, 4, 28, 10, 9, 45, 40, 67, 3, 62, 62, 14, 39, 46, 9, 63, 87, 31, 15, 67, 17, 81, 1, 91, 97, 53, 9, 66, 15, 34, 72, 68, 85, 20, 84, 36, 49]

```python
spisok = [9, 47, 78, 54, 92, 63, 59, 89, 96, 47, 95, 29, 99, 20, 4, 28, 10, 9, 45, 40, 67, 3, 62, 62, 14, 39, 46, 9, 63, 87, 31, 15, 67, 17, 81, 1, 91, 97, 53, 9, 66, 15, 34, 72, 68, 85, 20, 84, 36, 49]
dubl = [x for i, x in enumerate(spisok) if i != spisok.index(x)]
dubl1 = set(dubl)
print ("Повторяющиеся элементы: ", dubl1)
spisok = set(spisok)
print ("Уникальные значения: ", spisok - dubl1)

```

4. Написать функцию программу, принимающую 1 аргумент — число от 0 до 1000, и возвращающую True, если оно простое, и False - иначе

```python
number = int(input("Ввести число от 0 до 1000: "))
k = 0
for i in range(2, number // 2+1):
    if (number % i == 0):
        k = k+1
if (k <= 0):
    print("Число простое: ", True)
else:
    print("Число не является простым: ", False)
 
```
