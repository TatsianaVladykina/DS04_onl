```python
#Создание массива [1. 2. 3. 4. 5. 6. 7. 8. 9.] с помощью конструктора np.linspace
mas1 = np.linspace(1, 9, 9)
print(mas1)
```

```python
#Создание массива [9. 8. 7. 6. 5. 4. 3. 2. 1.] с помощью конструктора np.linspace
mas2 = np.linspace(9, 1, 9)
print(mas2)
```

```python
#Сравнение массивов между собой (операция >)
print(mas1 > mas2)
```

```python
#Сравнение массивов между собой (операция =)
print(mas1 == mas2)
```

```python
#Сравнение массивов между собой (операция <)
print(mas1 < mas2)
```

```python
#Выполнить поэлементное умножение массивов
mas3 = mas1 * mas2
print(mas3)
```

```python
#Найдите среднее значение массива полученного на шаге (print(mas1 * mas2))
mas3.mean()
```

```python
#Вычтите полученное среднее  (mas3.mean()), из массива (mas3 = mas1 * mas2)
```

```python
mas4 = mas3 - mas3.mean()
print(mas4)
```

```python
#Найдите скалярное произведение массивов/векторов(mas1 и mas2)
scalar_product = np.dot(mas1, mas2)
print(scalar_product)
```

```python
#Преобразуйте массивы (mas1, mas2) в двумерные массивы размером 3х3
resh_mas1 = mas1.reshape(3, 3)
resh_mas2 = mas2.reshape(3, 3)
print('Массив1 равен:\n', resh_mas1)
print('Массив2 равен:\n', resh_mas2)
```

```python
#Примените к полученным массивам функцию dot
scalar_product_resh_mas = np.dot(resh_mas1, resh_mas2)
print(scalar_product_resh_mas)
```
