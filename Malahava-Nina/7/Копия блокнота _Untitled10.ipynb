**ЗАДАЧА** ***Наибольший простой фактор***

Простые множители 13195: 5, 7, 13 и 29.

Какой самый большой простой делитель числа 600851475143?

```python
import math

def largest_simple_divider(n):
  i = 2
  while i * i <= n:
      if n % i:
        i += 1
      else:
        n //= i
  return n

print(largest_simple_divider(600851475143))

```

**ЗАДАЧА** ***Наименьшее кратное***

2520 - это наименьшее число, которое можно разделить на каждое из чисел от 1 до 10 без остатка.

Какое наименьшее положительное число делится без остатка на все числа от 1 до 20?

```python
def nod (a, b):
  while b > 0:
    a, b = b, a % b  # формула вычисления НОД
  return a


def nok (a, b):
  return a * b // nod(a, b) # НОК
  
c = 1
for i in range(2,21):  # перебираем наши делители
  c = nok(c, i)  # НОК с нашими делителями
print(c)
```
