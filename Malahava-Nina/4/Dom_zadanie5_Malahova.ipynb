# Обработка категориальных признаков

```python
# Метод, преобразует уникальные значения в столбце в числовые метки
h_cat_en, h_categories = X['ocean_proximity'].factorize()
```

```python
h_categories
h_cat_en
```

```python
# преобразует категориальные признаки в бинарные признаки, 
# путем создания новых столбцов для каждой уникальной категории и заполнения их значениями 0 и 1
from sklearn.preprocessing import OneHotEncoder
```

```python
# создаем объект
oh = OneHotEncoder()
```

```python
coded = oh.fit_transform(h_cat_en.reshape(-1,1))
```

```python
coded.toarray()
```

```python
proxy = pd.DataFrame(coded.toarray(), index = X.index, columns = ['1','2','3','4','5'])
```

```python
X = pd.concat([X, proxy], axis = 1)
```
