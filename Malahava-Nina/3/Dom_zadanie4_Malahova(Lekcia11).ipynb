```python
import numpy as np
import pandas as pd
```

```python
# Загрузка данных из файла.
data = np.loadtxt('linreg_4.txt')
print(data)
```

```python
# Отделение целевой переменной от независимых переменных.
x = data[:, :-1]  # независимые переменные (все столбцы, кроме последнего)
y = data[:, -1]   # целевая переменная (последний столбец)
```

```python
print(x)
```

```python
print(y)
```

```python
# Градиентный спуск - это метод оптимизации, 
# который используется для нахождения минимума функции, 
# изменяя ее параметры в направлении наиболее быстрого убывания функции. 
# В контексте линейной регрессии градиентный спуск может быть использован 
# для нахождения оптимальных значений коэффициентов модели.
```

```python
# Адаптируйте написанный на уроке код градиентного спуска для нахождения коэффициентов линейной модели вида 
# y = h0 + h1x1 + h2x2 + h3*x3.
# Загрузка данных по столбцам.
x1 = data[:, 0] # первый столбец (независимые переменные)
x2 = data[:, 1] # второй столбец (независимые переменные)
x3 = data[:, 2] # третий столбец (независимые переменные)
y = data[:, 3] # четвертый столбец (целевая переменная)
```

```python
print(x1)
```

```python
print(x2)
```

```python
print(x3)
```

```python
print(y)
```

```python
# Инициализация начальных значений параметров
h0 = 0
h1 = 0
h2 = 0
h3 = 0
```

```python
lr = 0.001  # скорость обучения
num_iters = 1000  # количество итераций градиентного спуска
m = len(y) # количество обучающих примеров, т.е. количество элементов в массиве у
```

```python
print(m)
```

```python
# Градиентный спуск для нахождения оптимальных параметров

for i in range(num_iters):
    y_pr = h0 + h1*x1 + h2*x2 + h3*x3
    grad_h0 = (1/m) * np.sum(y_pr - y)
    grad_h1 = (1/m) * np.sum((y_pr - y) * x1)
    grad_h2 = (1/m) * np.sum((y_pr - y) * x2)
    grad_h3 = (1/m) * np.sum((y_pr - y) * x3)
# Найдите коэффициенты h методом градиентного спуска   
    h0 = h0 - lr * grad_h0
    h1 = h1 - lr * grad_h1
    h2 = h2 - lr * grad_h2
    h3 = h3 - lr * grad_h3
    
```

```python
print('grad_h0 =', grad_h0)
print('grad_h1 =', grad_h1)
print('grad_h2 =', grad_h2)
print('grad_h3 =', grad_h3)
```

```python
# Вывод оптимальных параметров модели
print("h0 = {:.4f}".format(h0)) # выводит строку "h0 = " и значение переменной h0 с точностью до четырех знаков после запятой
print("h1 = {:.4f}".format(h1)) # выводит строку "h1 = " и значение переменной h1 с точностью до четырех знаков после запятой
print("h2 = {:.4f}".format(h2)) # выводит строку "h2 = " и значение переменной h2 с точностью до четырех знаков после запятой
print("h3 = {:.4f}".format(h3)) # выводит строку "h3 = " и значение переменной h3 с точностью до четырех знаков после запятой
```

```python
# построение графика для всех h (h0, h1, h2, h3)

import matplotlib.pyplot as plt

h0_values = []
h1_values = []
h2_values = []
h3_values = []

for i in range(num_iters):
    y_pr = h0 + h1*x1 + h2*x2 + h3*x3
    grad_h0 = (1/m) * np.sum(y_pr - y)
    grad_h1 = (1/m) * np.sum((y_pr - y) * x1)
    grad_h2 = (1/m) * np.sum((y_pr - y) * x2)
    grad_h3 = (1/m) * np.sum((y_pr - y) * x3)

# Найдите коэффициенты h методом градиентного спуска   
    h0 = h0 - lr * grad_h0
    h1 = h1 - lr * grad_h1
    h2 = h2 - lr * grad_h2
    h3 = h3 - lr * grad_h3

# Добавление нового значения коэффициента h в список
    h0_values.append(h0)
    h1_values.append(h1)
    h2_values.append(h2)
    h3_values.append(h3)

# Cоздание графиков
fig, axs = plt.subplots(4, 1, figsize=(15, 15), sharex=True)

axs[0].plot(h0_values)
axs[0].set_ylabel('h0')

axs[1].plot(h1_values)
axs[1].set_ylabel('h1')

axs[2].plot(h2_values)
axs[2].set_ylabel('h2')

axs[3].plot(h3_values)
axs[3].set_xlabel('Номер итерации')
axs[3].set_ylabel('h3')

plt.suptitle('Зависимость значений коэффициентов от номера итерации градиентного спуска')

plt.show()
```
