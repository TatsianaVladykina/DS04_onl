```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
start_directory = "d:/Works/Teach_ME_Skills/Материалы/ДЗ_Бруцкий-Стемпковский/DS3onl/Бруцкий-Стемпковский Николай/ДЗ_09_Бруцкий-Стемпковский"
path_to_set = start_directory + "/train_set"
path_to_save = start_directory + "/row_images"
```

```python
os.chdir(path_to_save)
list_of_new_images = os.listdir()

for im in list_of_new_images:
    os.remove(os.getcwd() + f"\\{im}")
```

```python
dict_of_categoriez = {}
```

```python
def save_images(cat, numbers):

    global path_to_set
    global path_to_save
    global start_directory

    dict_of_categoriez[cat] = []

    counter = 0
    file_obj = open(start_directory + "/annotation.txt", "r")

    while counter < numbers:
        string = file_obj.readline().replace(".", "").split()
        

        if int(string[cat]) == 100:

            os.chdir(path_to_set)
            image_obj = cv2.imread(string[0] + ".jpg")
            
            os.chdir(path_to_save)
            file_name = string[0] + ".jpg"
            cv2.imwrite(file_name, image_obj)
            counter += 1

            dict_of_categoriez[cat].append(file_name)
            

    file_obj.close()
```

Категории:

1. person
2. car
3. building
4. window
5. tree
6. sign
7. door
8. bookshelf
9. chair
10. table
11. keyboard
12. head

```python
save_images(2, 100)
```

```python
save_images(1, 20)
save_images(3, 20)
save_images(5, 20)
save_images(9, 20)
save_images(12, 20)
```
