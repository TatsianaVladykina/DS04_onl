Обработаем выбросы.

Для этого построим ящик с усами. Он показывает медиану, первый и третий квартили, усы отстоят на расстояние до 1,5(Q3 - Q1). Всё, что лежит за пределами усов - выбросы.

Атрибуты X3, X5, X6, Y имеют выбросы. Удалим сроки целиком...

any - удаление строки с хотя бы одним выбросом

```python
Q_1 = data.quantile(q = 0.25)
Q_3 = data.quantile(q = 0.75)
IQR = data.apply(sp.stats.iqr)
data_clean = data.loc[~((data < (Q_1-1.5*IQR))|(data > (Q_3+1.5*IQR))).any(axis = 1)]
(data.shape, data_clean.shape)
```

Выполним тест на нормальность. Тест выполняется при помощи критерия Шапиро-Уилка. Выдвигается гипотеза о том, что наше частное распеделение соответсвует нормальному распределению. Встроенный метод возвращает значение статистики и p-критерия для заданной выборки. Их смысл примерно свидетельствует о вероятности ошибочно назвать ненормальную выборку нормальной. В документации приводится, что при значении p-критерия больше 0,05 гипотезу о нормальности можно принимать.

```python
def normal_test(ser):
    stat, p = sp.stats.shapiro(ser)

    alpha = 0.05
    if p > alpha:
        return 'Похоже на нормальное распределение'
    else:
        return 'Не похоже на нормальное распределение'


for name in names:
    print(name, normal_test(data_clean[name]))
```
